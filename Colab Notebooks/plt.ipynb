```
data = pd.read_csv('005930.KS.csv')
```

```
data = data.dropna()
```

```
data.head()
```

```
date = data['Date']
price = data['Low']
price_1 = data['High']
```

```
fig1 = plt.figure(figsize=(16,8))
plt.plot(date[50:70], price[50:70], color = 'red', label = "LOW")
plt.plot(date[50:70], price_1[50:70], color = 'blue', label = "HIGH" )
plt.legend(loc = 'upper left')
plt.show()
```

```
fig1.savefig('fefef.png')
```

```
x = np.linspace(0, np.pi * 2, 50)
fig, axes = plt.subplots(figsize=(20,10))
axes.plot(x, np.sin(x), marker = '*', color = 'red', label = 'ss')
axes.plot(x, np.cos(x), label = 'sdd')
another = axes.twinx()
another.plot(date[50:70], price[50:70], color = 'red', label = "LOW")
another.plot(date[50:70], price_1[50:70], color = 'blue', label = "HIGH" )
axes.legend(loc = 'upper left')
another.legend(loc='lower left')
plt.show()
fig.savefig('sssd.png')
```
