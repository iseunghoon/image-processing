```
from sklearn import neighbors, datasets
import numpy as np

X = data.values
```

```
X.shape
```

```
X_set = X[:, 2:]
```

```
X_set.shape
```

```
from sklearn.preprocessing import scale, minmax_scale
Z = minmax_scale(X_set)
Z
```

```
data['target'] = data['diagnosis'].map({'B': 0, 'M' : 1})
data['target']
y_labels = np.array(data['target'])
y_labels
print(y_labels.shape)
```

```
from sklearn.model_selection import train_test_split
Z_train, Z_test, y_train, y_test = train_test_split(Z, y_labels, test_size=0.3)
```

```
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
knn.fit(Z_train, y_train)
pred = knn.pred = knn.predict(Z_test)
```

```
pred
```

```
pred.shape
```

```
y_test
```

```
from sklearn import metrics
print(metrics.confusion_matrix(y_test, pred))
```

```
print("정확도: {: .2f}".format(np.mean(y_test == pred)))
```

```
print("정확도: {: .2f}".format(knn.score(Z_test, y_test)))
```
