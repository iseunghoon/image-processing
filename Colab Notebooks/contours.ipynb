```
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('num.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)


ret, thres1 = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY)
thres1 = cv2.bitwise_not(thres1)

plt.imshow(cv2.cvtColor(thres1, cv2.COLOR_GRAY2RGB))
plt.show()

contours = cv2.findContours(thres1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0]
img = cv2.drawContours(img, contours, -1, (0,0,255),4)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

contour = contours[6]
x,y,w,h = cv2.boundingRect(contour)
image = cv2.rectangle(img, (x, y),(x + w, y + h),(0,0,255), 4)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```
hull = cv2.convexHull(contour)
img = cv2.drawContours(img, [hull], -1, (255,0,0), 4)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
