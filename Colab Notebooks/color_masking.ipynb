```
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```
Blue = 0
Green = 1
Red = 2
def select_color(img, color):
    other1 = (color + 1) % 3
    other2 = (color + 2) % 3
    c = img[:,:, other1] == 255
    img[c] = [0, 0, 0]
    c = img[:,:, other2] == 255
    img[c] = [0, 0, 0]
    c = img[:,:, color] < 170
    img[c] = [0, 0, 0]
    c = img[:,:,color] != 0 
    img[c] = [255,255,255]
    return img
```

```
num = cv2.imread('2.png')
```

```
red = select_color(num.copy(), Red)
```

```
plt.imshow(red)
```

```
num2 = cv2.imread('1.png')
```

```
red1 = select_color(num2.copy(), Red)
plt.imshow(red1)
```

```
red1_gray = cv2.cvtColor(red1, cv2.COLOR_BGR2GRAY)
```

```
plt.imshow(red1_gray)
```

```
ret, thres = cv2.threshold(red1_gray, 127, 255, cv2.THRESH_BINARY)
```

```
plt.imshow(thres, cmap='gray')
```

```
contours = cv2.findContours(thres, mode= cv2.RETR_LIST, method = cv2.CHAIN_APPROX_SIMPLE)[0]
```

```
contours
```

```
thres = cv2.drawContours(thres, contours=contours, contourIdx = -1, color=(255,0,0), thickness=2)
```

```
plt.figure(figsize=(20,10))
plt.imshow(thres, cmap='gray')
```

```
x,y,w,h = cv2.boundingRect(contours[2])
red1_gray = cv2.rectangle(red1_gray,pt1 = (x,y),pt2=(x+w,y+h), color=(255,0,0),thickness=2)
```

```
plt.figure(figsize=(20,10))
plt.imshow(red1_gray,cmap='gray')
```
