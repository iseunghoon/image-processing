```
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('dog.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
a = 4 ** 2
print(a)
```

```
size = 4
kernel = np.ones((size, size), np.float32) / (size ** 3)
print(kernel)
```

```
dst = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```
dst1 = cv2.blur(img, (4, 4))
plt.imshow(cv2.cvtColor(dst1, cv2.COLOR_BGR2RGB))
plt.show()
```

```
dst2 = cv2.GaussianBlur(img, (45, 45), 0)
plt.imshow(cv2.cvtColor(dst2, cv2.COLOR_BGR2RGB))
plt.show()
```

```
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thres1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(thres1, cv2.COLOR_GRAY2RGB))
plt.show()

contours = cv2.findContours(thres1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0]
img = cv2.drawContours(img, contours, -1, (0,0,255),4)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
