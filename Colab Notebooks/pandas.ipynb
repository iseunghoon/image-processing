열 추가

```
grade_dict = [{'name':'lee', 'midterm':90, 'final':100},
            {'name':'benjamin', 'midterm':85, 'final':95},
            {'name':'john', 'midterm':70, 'final':75},
            {'name':'chris', 'midterm':90, 'final':80} ]
```

```
grade_dict
```

```
grade_data = pd.DataFrame(grade_dict)
```

```
grade_data
```

```
grade_data['average'] = (grade_data['midterm'] + grade_data['final']) / 2
```

```
grade_data
```

```
grades = []
for row in grade_data['average']:
  if row > 90:
    grades.append('A')
  elif row > 80:
    grades.append('B')
  elif row > 70:
    grades.append('C')
  else:
    grades.append('D')
```

```
grades
```

```
grade_data['grade'] = grades
```

```
grade_data
```

```
def pass_fail(row):
  if row > 90:
    return 'GOOD'
  else:
    return 'NOT BAD'
```

```
grade_data.grade = grade_data.average.apply(pass_fail)
```

```
grade_data
```

```
date_list=[{'yyyy-mm-dd': '2019-02-15'},
           {'yyyy-mm-dd': '2020-03-20'},
           {'yyyy-mm-dd': '2020-06-30'}]
```

```
date_frame = pd.DataFrame(date_list)
```

```
date_frame
```

```
def extract_year(column):
  return column.split('-')[0]
  
```

```
date_frame['year'] = date_frame['yyyy-mm-dd'].apply(extract_year)
```

```
date_frame
```

```
def get_age(year, current_year):
  return current_year - int(year)
```

```
date_frame['age'] = date_frame['year'].apply(get_age, current_year= 2030)
```

```
date_frame
```

```
def get_introduce(age, prefix, suffix):
  return prefix + str(age) + suffix
```

```
date_frame['intro'] = date_frame['age'].apply(get_introduce, prefix = 'I am ', suffix = ' years old')
```

```
date_frame
```

```
def get_introduce2(row):
  return "I was born in"+str(row.year)+" I am "+str(row.age)+ " years old"
```

```
date_frame['intro2'] = date_frame.apply(get_introduce2, axis=1)
```
