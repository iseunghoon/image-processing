```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
Blue = 0
Green = 1
Red = 2
def select_color(img, color):
    other1 = (color + 1) % 3
    other2 = (color + 2) % 3
    c = img[:,:, other1] == 255
    img[c] = [0, 0, 0]
    c = img[:,:, other2] == 255
    img[c] = [0, 0, 0]
    c = img[:,:, color] < 170
    img[c] = [0, 0, 0]
    c = img[:,:,color] != 0 
    img[c] = [255,255,255]
    return img
```

```
def extract(img):
  chars = []
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  ret, thres = cv2.threshold(gray, 127, 255,0)
  
```

```
img = cv2.imread('2.png')
```

```
img_select = select_color(img, 2)
```

```
plt.imshow(img_select)
```

```
gray = cv2.cvtColor(img_select, cv2.COLOR_BGR2GRAY)
gray.shape
```

```
plt.imshow(gray, cmap='gray')
```

```
ret, gray_thres = cv2.threshold(gray, 127,255,0)
```

```
contours,_ = cv2.findContours(gray_thres, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```
len(contours)
```

```
img = cv2.drawContours(img_select, contours, -1 , (255,0,0), thickness=2)
```

```
plt.imshow(img)
```

```
x,y,w,h = cv2.boundingRect(contours[0])
print(w*h)
```
