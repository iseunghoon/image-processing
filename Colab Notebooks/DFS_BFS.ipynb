```
def dfs(graph, start):
  nodelist, edgelist = graph
  stack = [start]
  visited = []
  adjlist=[[]for node in nodelist]
  for edge in edgelist:
    adjlist[edge[0]].append(edge[1])
  while stack:
    current = stack.pop()
    for vertex in adjlist[current]:
      if not vertex in visited:
        stack.append(vertex)
    visited.append(current)
  return visited
```

```
dfs(graph, 1)
```

```
def bfs(graph, start):
  nodelist, edgelist = graph
  stack = [start]
  visited = []
  adjlist=[[]for node in nodelist]
  for edge in edgelist:
    adjlist[edge[0]].append(edge[1])
  while stack:
    current = stack.pop()
    for adj in adjlist[current]:
      if not adj in visited:
        stack.insert(0, adj)
    visited.append(current)
  return visited
```
