```
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```
img = cv2.imread('digits.png')
```

```
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```
plt.figure(figsize=(20,10))
plt.imshow(gray,cmap='gray')
```

```
cells = [np.hsplit(row, 100) for row in np.vsplit(gray, 50)]
x = np.array(cells)
```

```
x.shape
```

```
train = x[:,:].reshape(-1,400).astype(np.float32)
```

```
train.shape
```

```
k = np.arange(10)
```

```
train_labels = np.repeat(k, 500)
```

```
train_labels.shape
```

```
np.savez("data_set", train=train, train_labels=train_labels)
```

```
from sklearn import neighbors, datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(train, train_labels, test_size=0.1)  
```

```
knn = neighbors.KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
```

```
pred = knn.predict(x_test)
```

```
from sklearn import metrics
print(metrics.confusion_matrix(pred, y_test))
```

```
np.mean(pred == y_test)
```
