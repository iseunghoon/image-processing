```
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox

image_path = 'bycle_person.jpg' # 여기에는 테스트할 이미지의 경로 및 이름을 넣어주시면 됩니다. 
im = cv2.imread(image_path) # 이미지 읽기


# object detection (물체 검출)
bbox, label, conf = cv.detect_common_objects(im)

print(bbox, label, conf)

im = draw_bbox(im, bbox, label, conf) 


cv2.imwrite('result.jpg', im) # 이미지 쓰기
```

```
for i in range(5):
  temp = np.zeros((100,100))
  c = temp[:,:] == 0
  temp[c] = 255
  x = random.randint(0, 100) % 100
  y = random.randint(0,100) % 100
  x1 = x + 30
  y1 = y + 30
  if x1 > 100:
    x1 = 100
  elif y1 > 100:
    y1 = 100
  cv2.rectangle(temp, (x,y),(x1,y1),color=(0,0,0), thickness = 2)
  cv2.imwrite('result_{}.png'.format(i), temp)

```
